```python
file_path = 'merged.csv'

df = pd.read_csv(file_path, encoding='latin1')

# Specify the categories of interest
categories = [
    '4000-45000-10', '4000-44000-10', '4000-43000-10', '4000-42000-10', '4000-41000-10',
]

# Group by CATEGORY and count the non-null entries in ASSET_ID
category_counts = df.groupby('CATEGORY')['ASSET_ID'].count()

# Filter to include only the specified categories
category_counts_filtered = category_counts[category_counts.index.isin(categories)]

print(category_counts_filtered)
```
